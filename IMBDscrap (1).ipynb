```python
#Import the required modules.
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import csv

```

```python
#Access the HTML content from the webpage by assigning the URL and creating a soap object.
# Downloading imdb top 250 movie's data
result = requests.get('http://www.imdb.com/chart/top')
src= result.content
soup = BeautifulSoup(src, 'html5lib')
```

```python
movies = soup.find_all("td", {"class" : "titleColumn"})
```

```python
moviesTitle = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]

```

```python
moviesTitle
```

```python
rates = soup.find_all("td", {"class" : "ratingColumn imdbRating"})
```

```python
rates = [b.attrs.get('data-value')
        for b in soup.select('td.posterColumn span[name=ir]')]
```

```python
rates
```

```python
list = []
for index in range(0, len(movies)):
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    serial = movie[:len(str(index))-(len(movie))]
    data = {"serial": serial, "movie_title": movie_title, "rates": rates[index], "year": year}
    list.append(data)
```

```python
#list is filled with top IMBD movies along with their details.
for movie in list:
    print(movie['serial'], movie['movie_title'], movie['year'], movie['rates'])
 
 
```

```python
df = pd.DataFrame(list)
df.to_csv('imdb_top_250_movies.csv',index=False)
```
